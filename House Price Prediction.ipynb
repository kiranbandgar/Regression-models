```python
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
category_1_list = ["PoolQC","MiscFeature","Alley","Fence","FireplaceQu",
                      'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                     'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                      'BsmtFinType2',"MasVnrType",'MSSubClass']

for feature in category_1_list:
  data[feature] = data[feature].fillna('None')
```

```python
category_2_list = ['GarageYrBlt', 'GarageArea', 'GarageCars','BsmtFinSF1',
                      'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 
                      'BsmtHalfBath', "MasVnrArea"]
for feature in category_2_list:
  data[feature] = data[feature].fillna(0)
```

```python
category_3_list = ['MSZoning','Electrical','KitchenQual','Exterior1st','Exterior2nd',
                     'SaleType','Utilities']
for feature in category_3_list:
  data[feature] = data[feature].fillna(data[feature].mode()[0])
```

```python
category_4_list = ["LotFrontage"]
for feature in category_4_list:
  data[feature] = data[feature].fillna(data[feature].median())
```

```python
data['Functional'] = data['Functional'].fillna('Typ')
```

```python
datatype=data.select_dtypes(object)
```

```python
datatype
```

```python
datatypes_list=list(datatype.columns)
```

```python
datatypes_list
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse=False)
```

```python
data_encoded=ohe.fit_transform(datatype)
```

```python
data_encoded
```

```python
data_encoded_features=ohe.get_feature_names(datatypes_list)
data_encoded_features
```

```python
numerical_feature=data.select_dtypes(exclude=object)
```

```python
numerical_feature
```

```python
data_encoded_merged=pd.DataFrame(data_encoded,columns=data_encoded_features).astype(int)
data_encoded_merged.head()
```

```python
finaldata=pd.concat([data_encoded_merged, numerical_feature],axis=1)
```

```python
finaldata
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=finaldata.drop('SalePrice',axis=1)
```

```python
target=data[['SalePrice']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
p=model.fit(x_train,y_train).predict(x_test)
```

```python
p
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,p)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
p=model.fit(x_train,y_train).predict(x_test)
```

```python
r2_score(y_test,p)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
p=model.fit(x_train,y_train).predict(x_test)
```

```python
r2_score(y_test,p)
```
